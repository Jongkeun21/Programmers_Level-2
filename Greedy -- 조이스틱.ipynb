```python
class cal(object) :
    def __init__(self, name, dic) :
        self.name = name
        self.dic = dic
        
    def cal1(self, flag) :
        answer = 0
        pointer = 0
        result = ['A' for col in range(len(self.name))]

        while result != self.name :
            count = self.dic[self.name[pointer]]-self.dic[result[pointer]]
            result[pointer] = self.name[pointer]
            
            if count > 13 :
                count = 26-count

            answer += count
            answer += 1
            pointer += 1*flag

        return answer
    
    def cal2(self, flag) :
        answer = 0
        pointer = 0
        flag_ = True
        length = len(self.name)
        result = ['A' for col in range(length)]
        
        while result != self.name :
            if flag_ and pointer < length//2 + 1 :
                if self.name[pointer+1] == 'A' :
                    flag_ = False
                    flag *= -1
                
            count = self.dic[self.name[pointer]]-self.dic[result[pointer]]
            result[pointer] = self.name[pointer]

            if count > 13 :
                count = 26-count
                
            answer += count
            answer += 1
            pointer += 1*flag
            
        return answer
    
def solution(name) :
    alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    name = list(name)
    dic = {}
    
    for idx, alpha in enumerate(alphabets) :
        dic[alpha] = idx+1
        
    sample = cal(name, dic)
    num1 = sample.cal1(1)
    num2 = sample.cal1(-1)
    num3 = sample.cal2(1)
    num4 = sample.cal2(-1)
    
    return min(num1, num2, num3, num4)-1
```
